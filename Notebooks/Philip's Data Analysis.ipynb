```python
#Load dependencies
import os
import psycopg2
import pandas as pd
from pprint import pprint
import matplotlib.pyplot as plt
from scipy import stats
import statsmodels.api as sm 
import numpy as np
import psycopg2.extras
```

```python
#Setup connection
from sqlalchemy import create_engine
hostname = 'localhost'
username = 'postgres'
password = 'postgres'
database = 'energy_db'
connection = psycopg2.connect(host=hostname, user=username, password=password, dbname=database )
```

```python
statedata=pd.read_sql('SELECT * FROM state_data',con=connection)
stategreenhouse=pd.read_sql('SELECT * FROM state_greenhouse_emissions',con=connection)
statedata
```

```python
seriesgeneration=statedata.groupby(['state','year'])['generation_mwh'].sum()
generationtotal=statedata.loc[statedata['energy_source']=='Total']
generation=pd.DataFrame(seriesgeneration)
generation.reset_index(inplace=True)
generation.columns=['state','year','generation']
generation['year']=pd.to_numeric(generation['year'])
stategeneration=pd.merge(generationtotal,generation,on=['year','state'],how='outer')
stategeneration['calcgeneration']=stategeneration['generation']-stategeneration['generation_mwh']
del stategeneration['generation']
statetemp=pd.merge(stategeneration,stategreenhouse,on=['year','state'],how='outer')
del statetemp['consumption']
```
